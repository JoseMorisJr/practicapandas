```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ventas.csv")
df.head(15)
```

```python
#Obtener promedio de ventas mensuales por producto

df1 = df.groupby('codigo producto')['ventas mensuales'].agg(np.mean).reset_index()
df1 = df1.rename(columns={'ventas mensuales': 'promedio de ventas mensuales'})
df1.head()
```

```python
#Minimo Maximo y Promedio de Ventas por Producto

df1 = df.groupby('codigo producto')['ventas mensuales'].agg({'maximo':max, 'minimo': min, 'promedio':np.mean})
df1.head()
```

```python
#Minimo Maximo y Promedio de Ventas por Producto como filas
df3 = df1.stack().to_frame().reset_index()
df3.rename(columns = {'level_1' : 'magnitud', 0 : 'valor'}, inplace = True)
df3
```

```python
#Mismo con otro formato, uso pivot

#pivot(INDICE, COLUMNAS, VALORES)

df4 = df3.pivot('magnitud', 'codigo producto', 'valor')
df4.head()
```
