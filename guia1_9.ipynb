```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

internas = pd.read_csv("ventas_propias.csv")
externas = pd.read_csv("ventas_ext.csv")
```

```python
#  Se pide generar utilizando Pandas un dataframe que tenga el siguiente formato
#  (vehicle_model_id, ext_mean_price, ext_std_price, int_mean_price, int_std_price)

# - mean_price: precio promedio para ese vehículo.
# - std_price: desvío estándar del precio para ese vehículo.
# - los prefijo ext_ y int_ indicando que deben ser calculado sobre respectivamente datos
#   externos (los obtenido vía scraping) y datos internos (los de las transacciones).
```

```python
int_mean = internas.groupby('vehicle_model_id')['price'].agg(np.mean)
int_std = internas.groupby('vehicle_model_id')['price'].agg(np.std)

ext_mean = externas.groupby('vehicle_model_id')['price'].agg(np.mean)
ext_std = externas.groupby('vehicle_model_id')['price'].agg(np.std)

df = pd.merge(ext_mean, ext_std, on = 'vehicle_model_id', how = 'inner')
df = pd.merge(df, int_mean, on = 'vehicle_model_id', how = 'right')
df = pd.merge(df, int_std, on = 'vehicle_model_id', how = 'right')
```

```python
df.columns = ['ext_mean', 'ext_std', 'int_mean', 'int_std']
df.head()
```
